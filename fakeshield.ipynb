```python
# model_id = "CHELSEA234/llava-v1.5-7b-M2F2-Det"
model_id = "zhipeixu/fakeshield-v1-22b"
local_dir = "./llava_deepfake_model" # The directory where you want to save the model
```

```python
os.makedirs(local_dir, exist_ok=True)
```
